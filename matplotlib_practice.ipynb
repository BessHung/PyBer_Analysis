```python
# import denpendencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of string for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to the list of float as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Line chart_MATLAB Method - Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create the lable for the x and y axes
plt.xlabel('Month')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title('PyBer Fare by Month')
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, color ='g', marker = '*')
ax.legend('Boston')
ax.set_ylim(0,45)
ax.grid()
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
```

```python
# Bar chart_MATLAB Method
plt.bar(x_axis,y_axis, color='g',label='Boston')
plt.xlabel('Month')
plt.ylabel('Fare($)')
# Create a title.
plt.title("PyBer Fare by Month")
# Add a legend
plt.legend()
```

```python
# Create a Horizontal Bar Chart
plt.barh(x_axis,y_axis,color='magenta',label='Boston')
# invert the y-axis to have the months in ascending order
plt.gca().invert_yaxis()
plt.xlabel('Month')
plt.ylabel('Fare($)')
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Bar chart_object-oriented interface - Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis,label='Chicago')
ax.set_xlabel('Month')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color='cyan',label='Chicago')
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Month')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.legend()
```

```python
# Scatter plot_MATLAB Method
plt.plot(y_axis,x_axis,'o',color='r',label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
# increased size#1
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis,s=y_axis_larger,label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.legend()
```

```python
# increased size#2: new_list = [expression for item in list if conditional]
plt.scatter(x_axis,y_axis, s = [i*3 for i in y_axis], c='orange', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.legend()
```

```python
# Scatter plot_Object-Oriented Interface
fig, ax= plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],c='skyblue',
           edgecolors='black',alpha=0.8,label='Boston',linewidths=2 )
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
# Pie chart_MATLAB Method
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
# increase the size of pie chart
plt.subplots(figsize=(8,8))
# Assign 12 colors for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, labels=x_axis, explode=explode_values,autopct='%.1f%%',colors=colors,shadow=True)

plt.show()
```

```python
# Pie chart_Object-oriented interface
fig, ax = plt.subplots(figsize=(8,8))

explode= [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis, labels=x_axis,autopct='%.1f%%',shadow=True,colors=colors,startangle=90,explode=explode,counterclock=False)
ax.axis('equal')
plt.show()
```
